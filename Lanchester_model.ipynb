```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt

# Total Number of countries
N = 1

# Define the function for n as a function of k = k1/k2
def n(k):
    return N / (np.sqrt(k) + 1)

# Generate values for x
k_values = np.linspace(0.1, 2, 100)  # Avoid x = 0 for division by zero

# Calculate corresponding n values
n_values = n(k_values)

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(k_values, n_values, label=f'N={N}')
plt.xlabel('k1 / k2')
plt.ylabel('number of countries in coalition A')
plt.title('Number of countries in Coalition A vs. k1/k2')
plt.grid(True)

# Highlight the equilibrium point
equilibrium_x = 1
equilibrium_n = N / 2
plt.scatter(equilibrium_x, equilibrium_n, color='red', label='Equilibrium Point')
plt.annotate(f'Equilibrium Point: ({equilibrium_x}, {equilibrium_n})',
             xy=(equilibrium_x, equilibrium_n),
             xytext=(equilibrium_x + 0.2, equilibrium_n + 0.1),
             arrowprops=dict(facecolor='black', arrowstyle='->'),)

plt.legend()
plt.show()


# Print the values of k and n
for k, n_val in zip(k_values, n_values):
    print(f'k1/k2: {k:.2f}, n: {n_val:.2f}')
```
